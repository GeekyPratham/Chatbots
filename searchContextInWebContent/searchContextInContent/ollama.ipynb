```python
## Chatprompt tamplet -> role defined to llm model or what kind response we want from llm
from langchain_core.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate(
    [
        ("system","You are an expert AI engineer. Provide me answer based on qeustion"),
        ("user","{input}")
    ]
)
prompt
```

```python
## Chain A pipeline designed specifically for LLM workflows, often with:

# prompts
# models
# memory
# retrievers
# tools
# parsers

chain = prompt | llm
response = chain.invoke({
    "input" : "Can you tell me about langsmith?"
})
print(response)
```

```python

## Data ingestion -> from website we need to scrape the data 
from langchain_community.document_loaders import WebBaseLoader
loader = WebBaseLoader(
    "https://docs.langchain.com/langsmith/home",
    verify_ssl=False
)
print(loader)
```

```python
docs = loader.load()
""" 
loader.load() actual webpage par jaata hai aur:

HTML fetch karta hai
readable text nikalta hai
usko LangChain ke Document format me convert karta hai

"""
print(docs)
```

# load the data
# convert it into docs
# divide our text into chunks
# convert chunks into vector embedding
# store it into vector store db

```python
from  langchain_text_splitters import RecursiveCharacterTextSplitter
""" 
RecursiveCharacterTextSplitter --> Text ko manageable pieces me tod do, lekin possible ho toh sentences/paragraphs ka meaning bana rahe.
"""

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)

""" 

Problem: Hum pura webpage ek saath kyu nahi use karte?

Socho webpage me 20,000 characters ka content hai.

Agar tum pura content LLM ko de doge:

context window waste hoga
retrieval weak ho jayega
irrelevant information bahut jayegi
embeddings quality degrade ho sakti hai

Isliye hum content ko small meaningful pieces (chunks) me todte hain.

"""
documents = text_splitter.split_documents(docs)
```

```python
print(documents)
```

```python
documents
```

```python
## embedding it into vector
from langchain_community.embeddings import OllamaEmbeddings
embeddings = OllamaEmbeddings(
    model="gemma:2b"
)

""" 
gemma:2b ek chat/generative model hai, dedicated embedding model nahi.

Ollama embeddings ke liye better models hote hain jaise:

nomic-embed-text
mxbai-embed-large
all-minilm

"""
```

```python
## store it into vector db faiss
from langchain_community.vectorstores import FAISS #Facebook AI Similarity Search
vectorStore = FAISS.from_documents(documents,embeddings)
""" 
meaning of above line
1. Har document chunk ka embedding banata hai
2. Un embeddings ko index karta hai
3. Unko searchable vector store me daal deta hai

"""



"""  

Normal keyword search:

“debugging” word dhundega

Vector search:

“trace”, “evaluation”, “monitoring”, “observability” jaisi semantically similar cheeze bhi samajh lega

"""
```

```python
## querying in vector db
query = "LangSmith is a framework-agnostic platform for developing, debugging,"
result = vectorStore.similarity_search_with_score(query)
print(result)
```

```python
from langchain_ollama import ChatOllama
llm = ChatOllama(
    model="gemma:2b"
)
print(llm)
```

```python
## retrieval chain , document chain
from langchain_classic.chains.combine_documents import create_stuff_documents_chain ##Sare retrieved documents ko ek saath prompt me bhar dena
# this helps us to make a document chain
from langchain_core.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_template(
    """
    Answer the following question based on the the provided context:
    <context>
    {context}
    </context>

"""
)
document_chain = create_stuff_documents_chain(llm,prompt)

"""
Ye ek chain bana raha hai jo:

    documents ko lega
    unko format karega
    prompt me {context} me daalega
    LLM ko bhejega
    answer return karega

"""
document_chain
```

```python

from langchain_core.documents import Document


document_chain.invoke({
    "input":"LangSmith is a framework-agnostic platform for developing, debugging,",
    "context":[Document(page_content="LangSmith is a framework-agnostic platform for developing, debugging, and deploying AI agents and LLM applications. It helps you trace requests, evaluate outputs, test prompts, and manage deployments in one place. LangSmith works with any agent stack, whether you’re using an existing framework or building your own. Prototype locally, then move to production with integrated monitoring and evaluation to build more reliable AI agents.")]
})


""" 
ishme kya ho raha isme ham document kudh ka bhej rahe hai maan le ki document retrieve hone ke bad mila hai aur , ham apna input dia hai chain ko toh chain mera context lega document and input , llm ush document(retrieval chunk jab ham search karke nikale ge chunk) se search karke result dega

"""
```

However we want the documents first come from the retriever we just set up. That way, we can use the retriever to dynamically select the most relevant documents and pass those in for a given qeustion

```python
## ham log upar me document(context ) bhej rahe the wah document ke form me hard coded tha
## magar actual me retrive document (context, chunk ) ke basis par bhejna hota hai
```

```python
## Retriever
""" 
input come -> goes to retrieva(act as interface) -> bring the relevant data from the vectorStore db .

so we have to convert vectorStore db into retriever

and we even dont have to do similarity search retriever does for us.
"""
vectorStore
```

```python
retriever = vectorStore.as_retriever() # convert vectorStore db into retriever
```

```python
from langchain_classic.chains import create_retrieval_chain
retrieval_chain = create_retrieval_chain(retriever,document_chain)
## this take out relevant context from the db and fill the context into the prompt and then we can invoke prompt with llm
```

```python
retrieval_chain
```

```python
## get the response from the llm
response = retrieval_chain.invoke(
    {
        "input":"LangSmith is a framework-agnostic platform for developing, debugging,",
    }
)
```
